### 1. Calculate the average rental duration (in days) for each film:

```python
import pandas as pd
import sqlalchemy as alch
from getpass import getpass
password = getpass()
dbName = "sakila"
connectionData=f"mysql+pymysql://root:{password}@localhost/{dbName}"
engine = alch.create_engine(connectionData)
```

```python
query = '''SELECT film.title, film.rental_duration, AVG(DATEDIFF(rental.return_date, rental.rental_date)) AS avg_rental_duration
FROM FILM
JOIN INVENTORY ON FILM.FILM_ID = INVENTORY.FILM_ID
JOIN RENTAL ON INVENTORY.INVENTORY_ID = RENTAL.INVENTORY_ID
GROUP BY film.title, film.rental_duration; '''

pd.read_sql_query(query, engine)
```

###  2. Calculate the average payment amount for each staff member:

```python
query = '''
SELECT staff_id, AVG(amount) AS avg_payment_amount
FROM PAYMENT 
GROUP BY staff_id; '''


pd.read_sql_query(query, engine)
```

### 3. Calculate the total revenue for each customer, showing the running total within each customer's rental history:

```python
query = '''
select payment.customer_id, rental.rental_id, rental.rental_date, payment.amount, sum(payment.amount) over (partition by payment.customer_id order by rental.rental_date) as running_total
from payment 
join rental on payment.rental_id = rental.rental_id;
'''

pd.read_sql_query(query, engine)
```

### 4. Determine the quartile for the rental rates of films:

```python
query = ''' 
SELECT film.title, film.rental_rate, NTILE(4) over (PARTITION BY film.rental_rate) AS QUARTILE
from FILM;
''' 

pd.read_sql_query(query, engine)
```

### 6. Calculate the rank of customers based on their rental counts:

```python
query = '''
SELECT rental.customer_id, COUNT(rental.customer_id) AS rental_count, RANK() OVER (ORDER BY COUNT(rental.customer_id) desc) AS rental_count_rank 
FROM RENTAL
GROUP BY rental.customer_id;
''' 

pd.read_sql_query(query, engine)
```

### 7. Calculate the running total of revenue per day for the 'Family' film category:

```python
query = '''
SELECT film.title, rental.rental_date, payment.amount, sum(payment.amount) OVER (PARTITION BY RENTAL.RENTAL_DATE) AS daily_revenue
FROM CATEGORY 
JOIN FILM_CATEGORY ON CATEGORY.CATEGORY_ID = FILM_CATEGORY.CATEGORY_ID
JOIN FILM ON FILM_CATEGORY.FILM_ID = FILM.FILM_ID 
JOIN INVENTORY ON FILM.FILM_ID = INVENTORY.FILM_ID
JOIN RENTAL ON INVENTORY.INVENTORY_ID = RENTAL.INVENTORY_ID
JOIN PAYMENT ON RENTAL.RENTAL_ID = PAYMENT.RENTAL_ID
WHERE CATEGORY.NAME = "FAMILY";
''' 

pd.read_sql_query(query, engine)
```

### 9. Calculate the difference in days between each rental and the previous rental for each customer:

```python
query = ''' 
SELECT rental.customer_id, rental.rental_id, rental.rental_date, lag(rental.rental_date) over (partition by customer_id order by rental_date) as previous_rental_date, 
datediff (rental_date, lag(rental_date) over (partition by customer_id order by rental_date)) as days_between_rentals
FROM RENTAL;
''' 

pd.read_sql_query(query, engine)
```
