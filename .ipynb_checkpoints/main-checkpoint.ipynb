### 3. Calculate the total revenue for each customer, showing the running total within each customer's rental history:

The ORDER BY clause inside the OVER() function is used to define the order in which the rows are processed for the window function. This is crucial when you're calculating running totals, moving averages, or any other cumulative or window-based metric.

Here’s why and when you might use it:

#### Why Use ORDER BY Inside OVER()?

Running Totals: When calculating a running total, you need to define the order in which the rows should be added up. For example, if you are calculating a running total of sales over time, you would order the rows by the date.

Ranking: If you are using window functions to rank items (using ROW_NUMBER(), RANK(), DENSE_RANK(), etc.), you need to define the order in which the items should be ranked.

Moving Averages: When calculating a moving average, the order of the rows determines which rows are included in the average for each row.

#### When to Use ORDER BY Inside OVER()?

Always for Running Totals: If you are calculating a running total, you should always use ORDER BY to define the order of accumulation.

For Ranking and Window Functions: Whenever you are using a window function that depends on the order of the rows (like LEAD(), LAG(), FIRST_VALUE(), LAST_VALUE(), etc.), you should use ORDER BY.

Sometimes for Aggregations: For windowed aggregations (SUM(), AVG(), etc.), use ORDER BY if you want a cumulative or moving calculation. If you just want a total over a partition without considering order, you can omit the ORDER BY.


````python
query = '''
    SELECT cu.customer_id, r.rental_id, r.rental_date, py.amount, sum(py.amount) OVER(PARTITION BY cu.customer_id ORDER BY rental_date) AS running_total
    FROM customer cu
    JOIN rental r ON cu.customer_id = r.customer_id
    JOIN payment py ON r.rental_id = py.rental_id
    ORDER BY cu.customer_id, r.rental_id;

'''
pd.read_sql_query(query, engine)

````

### 7. Calculate the running total of revenue per day for the 'Family' film category:

````python
query = '''
   
SELECT title as film_category, rental_date, amount,
SUM(amount) OVER (PARTITION BY rental_date) AS daily_revenue
FROM film
    JOIN inventory
        ON film.film_id = inventory.film_id
    JOIN rental
        ON inventory.inventory_id = rental.inventory_id
    JOIN payment
    ON rental.rental_id = payment.rental_id
WHERE rating = 'G';
'''
pd.read_sql_query(query, engine)



````

### 9. Calculate the difference in days between each rental and the previous rental for each customer:

````python
query = ''' 
    SELECT customer_id, rental_id,rental_date , DATEDIFF(rental_date, LAG(rental_date) OVER (PARTITION BY customer_id ORDER BY rental_date)) AS days_between_rentals
    FROM rental
    ORDER BY customer_id, rental_date;
  '''
pd.read_sql_query(query, engine)
````

````python
query = ''' 
    SELECT customer_id, rental_id, rental_date, LAG(rental_date) OVER (PARTITION BY customer_id ORDER BY rental_date) AS previous_rental_date,
    DATEDIFF(rental_date, LAG(rental_date) OVER (PARTITION BY customer_id ORDER BY rental_date)) AS days_between_rentals
    FROM rental
    ORDER BY customer_id, rental_date;
    
    '''
pd.read_sql_query(query, engine)
````
